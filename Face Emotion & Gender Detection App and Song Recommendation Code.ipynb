```python
import cv2
from deepface  import DeepFace
import tkinter as tk
from tkinter import *
from playsound import playsound
from pygame import mixer

def stop_music():
    mixer.music.stop()
   
def displaycamera():
   
        global gender
        global dominant_emotion

        faceCascade=cv2.CascadeClassifier(cv2.data.haarcascades+"haarcascade_frontalface_default.xml")
        cap = cv2.VideoCapture(1)
        if not cap.isOpened():
            cap = cv2.VideoCapture(0)
        if not cap.isOpened():
            raise IOError("Cannot Open Webcam")

        while True:
                ret,frame = cap.read()
                result = DeepFace.analyze(frame, actions=['emotion'],enforce_detection=False)
               
                gender_result = DeepFace.analyze(frame, actions=['gender'],enforce_detection=False)
               
                gender = gender_result['gender']
                dominant_emotion = result['dominant_emotion']
                               
                gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
                faces= faceCascade.detectMultiScale(gray,1.1,4)
                for (x,y,w,h) in faces:
                    cv2.rectangle(frame, (x,y), (x+w,y+h), (0,255,0), 2)
                font =cv2.FONT_HERSHEY_SIMPLEX
                cv2.putText(frame,result['dominant_emotion'],(50,50),font,3,(0,0,255),2,cv2.LINE_4)
                cv2.imshow("Demo Video",frame)

                if cv2.waitKey(2) & 0xFF==ord('q'):
                    break
             
        label2 = tk.Label(root, text=result['dominant_emotion'],bg='orange')
        canvas1.create_window(300, 380, window=label2)
       
        label3 = tk.Label(root, text='Dominant Emotion :',fg="#0000FF",bg='yellow')
        canvas1.create_window(150, 380, window=label3)
       
       
               
        button2 = tk.Button (root, text='Stop Music',command=stop_music, bg='orange')
        canvas1.create_window(250, 450, window=button2)
   
        button = tk.Button (root, text=gender, bg='orange')
        canvas1.create_window(200, 350, window=button)

        label2 = tk.Label(root, text=' Gender : ')
        canvas1.create_window(150, 350, window=label2)
     
        cap.release()
        cv2.destroyAllWindows()
       
def sel():
       
    selectedText = "You selected the option " + str(var.get())
    label_radio = tk.Label(root, text=selectedText)
    canvas1.create_window(150, 250, window=label_radio)
   
    if str(var.get())=='1' :
     
       
         # Playing Music File
        if gender=='Man':
             sound_file ='Male.mp3'
        elif gender=='Women':
             sound_file ='Female.mp3'
        mixer.init()
        mixer.music.load(sound_file)
        mixer.music.play()
     
    if str(var.get())=='2' :
     
         # Playing Music File
        if dominant_emotion =='happy':
             sound_file ='Happy.mp3'
        elif dominant_emotion=='sad':
             sound_file ='Sad.mp3'
        elif dominant_emotion=='angry':
             sound_file ='Angry.mp3'
        elif dominant_emotion=='neutral':
             sound_file ='Neutral.mp3'
        mixer.init()
        mixer.music.load(sound_file)
        mixer.music.play()
       
   

root= tk.Tk()
canvas1 = tk.Canvas(root, width = 500, height = 500)
canvas1.pack()

label_title = tk.Label(root, text=' Emotion Detection and Song Recommendation : ', anchor='center', font=("Arial", 15),
                       bg='#fff',
                       fg='#f0f',)
canvas1.create_window(250, 20, window=label_title)

label1 = tk.Label(root, text=' Name : ')
canvas1.create_window(100, 100, window=label1)
entry1 = tk.Entry (root)
canvas1.create_window(270, 100, window=entry1)


button1 = tk.Button (root, text='Capture Emotion',command=displaycamera, bg='orange')
canvas1.create_window(250, 150, window=button1)

var=IntVar(0)
label_radio = tk.Label(root, text=' Play Music By : ')
canvas1.create_window(150, 200, window=label_radio)

R1 = Radiobutton(root, text="Gender", variable=var, value=1,command=sel,background = "light blue")
R1.place(x=250,y=200)
R2 = Radiobutton(root, text="Emotion", variable=var, value=2,command=sel,background = "light blue")
R2.place(x=250,y=220)

root.mainloop()
```
